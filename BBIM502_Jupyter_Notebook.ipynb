```python
# Name: Arshdeep Singh
# Filename: BBIM502_ASSESSMENT_2
# Description: Application for “LANGHAM HOTELS”
# Date: 14-07-2024
try:
    import os  # Importing os module for file and directory operations
    import datetime  # Importing datetime module to handle date and time
    
    # Function to add a new room
    def Add_Room(rooms):
        try:
            # Input Room Number, Room Type, Room Prize from the User
            room_number = int(input("Enter Room Number: "))
            room_type = input("Enter Room Type: ")
            room_prize = float(input("Enter Room Prize (per day): $"))
        
            # If User Not Input Room Number or Room Type or Room Prize Raise Value Error
            if not room_number or not room_type or not room_prize:
                raise ValueError("All fields are required")

            # Set availability equal to available
            availability = "available"
            # APPEND room_number, room_type, room_prize, availability to "rooms" LIST
            rooms.append({
                "room_number": room_number,
                "room_type": room_type,
                "room_prize": room_prize,
                "availability": availability
            })
            print("Room Added Successfully.")

        # EXCEPT raise Value Error and PRINT the Message
        except ValueError as e:
            print(f"Value Error: {e}")
        except TypeError as e:
            print(f"Type Error: {e}")

    # Function to delete a room
    def Delete_Room(rooms):
        try:
            # Input room_number from the User
            room_number = int(input("Enter Room Number to delete: "))

            # If User not enter room_number THEN RAISE Value Error
            if not room_number:
                raise ValueError("Room Number Cannot Be Empty")

            # Check room in rooms LIST
            for room in rooms:
                if room["room_number"] == room_number: 
                    rooms.remove(room)  # REMOVE room from the rooms LIST
                    print("Room Deleted Successfully.")
                    return
        
            raise KeyError("Room Not Found")

        # EXCEPT raise Value Error and Print the Message
        except ValueError as e:
            print(f"Value Error: {e}")
        # EXCEPT raise Key Error and Print the Message
        except KeyError as e:
            print(f"Key Error: {e}")

    # Function to display all room details
    def Display_Rooms_Details(rooms):
        try:

            # IF rooms LIST is EMPTY THEN RAISE Index Error
            if not rooms:
                raise IndexError("No Rooms Found In The Inventory")

            # Check room in rooms LIST
            for room in rooms:
                # PRINT the room details
                print("-----Room Details-----")
                print(f"Room Number: {room['room_number']}") 
                print(f"Room Type: {room['room_type']}") 
                print(f"Room Prize: {room['room_prize']}")
                print(f"Status: {room['availability']}")

        # EXCEPT RAISE Index Error and PRINT the MESSAGE
        except IndexError as e:
            print(f"Index Error: {e}")

        # EXCEPT HANDLE Syntax Error and PRINT the MESSAGE
        except SyntaxError as e:
            print(f"Syntax Error: {e}")

    # Function to allocate a room to a guest
    def Allocate_Room(rooms, allocation):
        try:
            # INPUT room_number from the User
            room_number = int(input("Enter Room Number to Allocate: "))

            # IF User not enter the room_number THEN RAISE VALUE ERROR
            if not room_number:
                raise ValueError("Room Number Cannot Be Empty")

            # Check room in rooms LIST 
            for room in rooms:
                # Check room_number is equal to room_number added in rooms LIST and 
                # Availability is Equal to available 
                if room["room_number"] == room_number and room["availability"] == "available":
                    # INPUT guest ID 
                    guest_id = input("Enter Guest's Passport/Driving License ID: ")
                    # INPUT guest name
                    guest_name = input("Enter Guest's Name: ")
                    # INPUT guest contact
                    guest_contact = input("Enter Guest's Contact Information: ")
                    # INPUT allocating date
                    allocating_date = input("Enter Allocating Date (DD-MM-YYYY): ")
                    # INPUT allocating time
                    allocating_time = input("Enter Allocating Time (HH:MM): ")

                    # IF user not enter guest_id or guest_name or guest_contact or allocating_date or 
                    # allocating_time THEN RAISE Value Error
                    if not guest_id or not guest_name or not guest_contact or not allocating_date or not allocating_time:
                        raise ValueError("All Fields Are Required")

                    # APPEND guest_id, guest_name, guest_contact, allocating_date, allocating_time
                    # to "allocation" dictionary
                    allocation[room_number] = {
                        "guest_id": guest_id,
                        "guest_name": guest_name,
                        "guest_contact": guest_contact,
                        "allocating_date": allocating_date,
                        "allocating_time": allocating_time,
                    }
                    room["availability"] = "allocated"
                    print("Room Allocated Successfully.")
                    return

            # Raise Key Error and PRINT MESSAGE
            raise KeyError("Room Not Found Or Already Allocated")

        # EXCEPT RAISE Value Error and PRINT MESSAGE
        except ValueError as e:
            print(f"Value Error: {e}")

        # EXCEPT RAISE Key Error and PRINT MESSAGE
        except KeyError as e:
            print(f"Key Error: {e}")

    # Function to display allocation details
    def Display_Allocation_Details(allocation):
        try:
            # IF allocation dictionary is EMPTY THEN RAISE Index Error
            if not allocation:
                raise IndexError("No Allocations Found In The Inventory")
            # Check FOR room_number and details in the allocation dictionary
            for room_number, details in allocation.items():
                # PRINT room_number
                print(f"Room Number: {room_number}")
                # PRINT guest_id
                print(f"Guest ID: {details['guest_id']}")
                # PRINT guest_name
                print(f"Guest Name: {details['guest_name']}")
                # PRINT guest_contact
                print(f"Guest Contact: {details['guest_contact']}")
                # PRINT allocating_date
                print(f"Allocating Date: {details['allocating_date']}")
                # PRINT allocating_time
                print(f"Allocating Time: {details['allocating_time']}")

        # EXCEPT RAISE Index Error and PRINT MESSAGE
        except IndexError as e:
            print(f"Index Error: {e}")

    # Function to bill and deallocate a room
    def Billing_De_Allocation(rooms, allocation):
        try:
            # INPUT room_number from the User
            room_number = int(input("Enter Room Number To Deallocate: "))

            # IF User not enter room_number THEN RAISE Value Error
            if not room_number:
                raise ValueError("Room Number Cannot be Empty.")

            # Check IF room_number in allocation 
            if room_number in allocation:
                # INPUT duration_of_stay from the User
                duration_of_stay = int(input("Number of Days Stay: "))

                # IF User input duration_of_stay less than 0 or zero then RAISE Value Error
                if duration_of_stay <= 0:
                    raise ValueError("Duration of Stay must be greater than zero")
            
                # Get room_prize from the rooms LIST
                room_prize = next((room["room_prize"] for room in rooms if room["room_number"] == room_number), None)

                # IF no room_prize THEN RAISE Key Error
                if room_prize is None:
                    raise KeyError("Room Prize not found for the room")
                    
                # OverFlowError Testing
                if duration_of_stay >= 10**2:
                    raise OverFlowError("Simulated overflow error for testing purposes")
                    
                # Bill Amount = Duration of Stay Multiply Room Prize
                bill_amount = (duration_of_stay * float(room_prize))

                # FOR room in rooms LIST
                for room in rooms:
                    # IF room_number is equal to room_number
                    if room["room_number"] == room_number:
                        # SET availability equal to available
                        room["availability"] = "available"
                        break
                # DELETE room from the allocation dictionary
                del allocation[room_number]

                # PRINT Bill Details,
                print("Room Deallocated Successfully.")
                print("------Bill Details------")
                print(f"Room Number: {room_number}")
                print(f"Number of Days Stay: {duration_of_stay}")
                print(f"Room Prize: ${room_prize} per day")
                print(f"------Bill Amount: ${bill_amount}------")

            # ELSE RAISE Key Error and PRINT the MESSAGE
            else:
                raise KeyError("Room Not Found In Allocation Records.")

        # EXCEPT RAISE Value Error and PRINT MESSAGE
        except ValueError as e:
            print(f"Value Error: {e}")
        
        # EXCEPT RAISE Key Error and PRINT MESSAGE
        except KeyError as e:
            print(f"Key Error: {e}")
        
        # EXCEPT RAISE Key Error and PRINT MESSAGE
        except OverflowError as e:
            print(f"Overflow Error: {e}")
        
        # EXCEPT RAISE ZeroDivisionError and PRINT MESSAGE
        except ZeroDivisionError as e:
            print(f"Zero Division Error: {e}")

    # Function to save room allocation details to a file
    def Save_Room_Allocation_Details_To_File(filename, allocation):
        try:
            # WITH OPEN FILE in APPEND Mode
            with open(filename, 'a') as file:
                # WRITE room allocation details in the file
                for room_number, details in allocation.items():
                    file.write(f"{room_number},{details['guest_id']},{details['guest_name']},{details['guest_contact']},{details['allocating_date']},{details['allocating_time']}\n")
            print("Allocation Details Saved To File Successfully.")

        # EXCEPT RAISE IO Error and PRINT MESSAGE
        except IOError as e:
            print(f"IO Error: {e}")

    # Function to display details from a file
    def Display_Room_Allocation_Details_From_File(filename):
        try:
            # IF File not Found at the location THEN RAISE FILENOTFOUNDERROR
            if not os.path.isfile(filename):
                raise FileNotFoundError("File not found. Please add data first.")

            # WITH OPEN FILE IN READ MODE 
            with open(filename, 'r') as file:
                lines = file.readlines()

            # IF no data in the file THEN RAISE Lookup Error
            if not lines:
                raise LookupError("No data found in the file.")
        
            # Display details from the file, room_number, guest_id, guest_name, 
            # guest_contact, allocating_date, allocating_time 
            for line in lines:
                data = line.strip().split(',')
                if len(data) == 6:  # Allocation details
                    print("----Room Allocation Details----")
                    print(f"Room Number: {data[0]}")
                    print(f"Guest ID: {data[1]}")
                    print(f"Guest Name: {data[2]}")
                    print(f"Guest Contact: {data[3]}")
                    print(f"Allocating Date: {data[4]}")
                    print(f"Allocating Time: {data[5]}")
                    print("--------------------------------")

        # EXCEPT RAISE File Not Found Error and PRINT MESSAGE
        except FileNotFoundError as e:
            print(f"FileNotFoundError: {e}")

        # EXCEPT RAISE Lookup Error and PRINT MESSAGE
        except LookupError as e:
            print(f"LookupError: {e}")

        # EXCEPT RASIE IO Error and PRINT MESSAGE
        except IOError as e:
            print(f"IOError: {e}")

        except NameError as e:
            print(f"Name Error: {e}")

    # Function to backup data to another file
    def Backup_File(filename):
        try:
            # IF File Not Found Then Raise File Not Found Error
            if not os.path.isfile(filename):
                raise FileNotFoundError("File not found.")

            # Create backup_filename 
            backup_filename = f"LHMS_764708022_Backup_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

            # WITH OPEN filename in READ MODE
            with open(filename, 'r') as file:
                # READ content of the "filename"
                content = file.read()
                #  EOFError by reading an empty file
                if content == "":
                    raise EOFError("Unexpected end of file while reading.")

            # WITH OPEN backup_filename in APPEND MODE
            with open(backup_filename, 'a') as backup_file:
                # WRITE content of the "filename" to "backup_filename"
                backup_file.write(content)
            # WITH OPEN filename in WRITE MODE 
            with open(filename, 'w'):  # Clear filename content
                pass
            # PRINT BACKUP FILE NAME AND MESSAGE
            print(f"Backup Created Successfully: {backup_filename}")

        # EXCEPT RAISE File Not Found Error and PRINT MESSAGE
        except FileNotFoundError as e:
            print(f"File Not Found Error: {e}")
        
        # EXCEPT RAISE IO Error and PRINT MESSAGE
        except IOError as e:
            print(f"IO Error: {e}")
        
        # EXCEPT RAISE EOF Error and PRINT MESSAGE
        except EOFError as e:
            print(f"EOF Error: {e}")

    # Main function to handle menu options
    def main():
        try:
            # Create EMPTY rooms LIST
            rooms = []
            # Create EMPTY room Dictionary
            allocation = {}
            # Create filename and txt file
            filename = "LHMS_764708022.txt"

            while True:
                # PRINT HOTEL MENU OPTIONS
                print("\n----- LANGHAM HOTEL MANAGEMENT SYSTEM -----")
                print("              --- MENU ---                 ")
                print("1. Add Room")
                print("2. Delete Room")
                print("3. Display Rooms Details")
                print("4. Allocate Room")
                print("5. Display Room Allocation Details")
                print("6. Billing & De-Allocation")
                print("7. Save Room Allocation Details To File")
                print("8. Display Room Allocation Details From File")
                print("9. Backup Data")
                print("0. Exit Application")

                # PROMPT the user to Input Choice
                choice = input("Enter your choice: ")

                # IF choice equal to 1, CALL Add_Room FUNCTION
                if choice == '1':
                    Add_Room(rooms)
                # IF choice equal to 2, CALL Delete_Room FUNCTION
                elif choice == '2':
                    Delete_Room(rooms)
                # IF choice equal to 3, CALL Display_Room_Details FUNCTION
                elif choice == '3':
                    Display_Rooms_Details(rooms)
                # IF choice equal to 4, CALL Allocate_Room FUNCTION
                elif choice == '4':
                    Allocate_Room(rooms, allocation)
                # IF choice equal to 5, CALL Display_Allocation_Details FUNCTION
                elif choice == '5':
                    Display_Allocation_Details(allocation)
                # IF choice equal to 6, CALL Billing_De_Allocation FUNCTION
                elif choice == '6':
                    Billing_De_Allocation(rooms, allocation)
                # IF choice equal to 7, CALL Save_Room_Allocation_Details_To_File FUNCTION
                elif choice == '7':
                    Save_Room_Allocation_Details_To_File(filename, allocation)
                # IF choice equal to 8, CALL Display_Room_Allocation_Details_From_File FUNCTION
                elif choice == '8':
                    Display_Room_Allocation_Details_From_File(filename)
                # IF choice equal to 9, CALL Backup_File FUNCTION
                elif choice == '9':
                    Backup_File(filename)
                # IF choice equal to 0, Exiting The Application BREAK
                elif choice == '0':
                    print("Exiting The Application....")
                    break
                # ELSE PRINT "Invalid choice..." MESSAGE
                else:
                    print("Invalid Choice. Please Try Again.")
    
        except (SyntaxError, ValueError, ZeroDivisionError, IndexError, NameError, TypeError, OverflowError, IOError, ImportError, EOFError, FileNotFoundError) as e:
            print(f"An Error Occurred: {e}")


    # CALLING MAIN FUNCTION
    if __name__ == "__main__":
        main()

except ImportError as e:
    print(f"Import Error: {e}")
```

```python
    # Function to add a new room
    def Add_Room(rooms):
        try:
            # Input Room Number, Room Type, Room Prize from the User
            room_number = int(input("Enter Room Number: "))
            room_type = input("Enter Room Type: ")
            room_prize = float(input("Enter Room Prize (per day): $"))
        
            # If User Not Input Room Number or Room Type or Room Prize Raise Value Error
            if not room_number or not room_type or not room_prize:
                raise ValueError("All fields are required")

            # Set availability equal to available
            availability = "available"
            # APPEND room_number, room_type, room_prize, availability to "rooms" LIST
            rooms.append({
                "room_number": room_number,
                "room_type": room_type,
                "room_prize": room_prize,
                "availability": availability
            })
            print("Room Added Successfully.")

        # EXCEPT raise Value Error and PRINT the Message
        except ValueError as e:
            print(f"Value Error: {e}")
        except TypeError as e:
            print(f"Type Error: {e}")

    # Function to allocate a room to a guest
    def Allocate_Room(rooms, allocation):
        try:
            # INPUT room_number from the User
            room_number = int(input("Enter Room Number to Allocate: "))

            # IF User not enter the room_number THEN RAISE VALUE ERROR
            if not room_number:
                raise ValueError("Room Number Cannot Be Empty")

            # Check room in rooms LIST 
            for room in rooms:
                # Check room_number is equal to room_number added in rooms LIST and 
                # Availability is Equal to available 
                if room["room_number"] == room_number and room["availability"] == "available":
                    # INPUT guest ID 
                    guest_id = input("Enter Guest's Passport/Driving License ID: ")
                    # INPUT guest name
                    guest_name = input("Enter Guest's Name: ")
                    # INPUT guest contact
                    guest_contact = input("Enter Guest's Contact Information: ")
                    # INPUT allocating date
                    allocating_date = input("Enter Allocating Date (DD-MM-YYYY): ")
                    # INPUT allocating time
                    allocating_time = input("Enter Allocating Time (HH:MM): ")

                    # IF user not enter guest_id or guest_name or guest_contact or allocating_date or 
                    # allocating_time THEN RAISE Value Error
                    if not guest_id or not guest_name or not guest_contact or not allocating_date or not allocating_time:
                        raise ValueError("All Fields Are Required")

                    # APPEND guest_id, guest_name, guest_contact, allocating_date, allocating_time
                    # to "allocation" dictionary
                    allocation[room_number] = {
                        "guest_id": guest_id,
                        "guest_name": guest_name,
                        "guest_contact": guest_contact,
                        "allocating_date": allocating_date,
                        "allocating_time": allocating_time,
                    }
                    room["availability"] = "allocated"
                    print("Room Allocated Successfully.")
                    return

            # Raise Key Error and PRINT MESSAGE
            raise KeyError("Room Not Found Or Already Allocated")

        # EXCEPT RAISE Value Error and PRINT MESSAGE
        except ValueError as e:
            print(f"Value Error: {e}")

        # EXCEPT RAISE Key Error and PRINT MESSAGE
        except KeyError as e:
            print(f"Key Error: {e}")



    # Function to bill and deallocate a room
    def Billing_De_Allocation(rooms, allocation):
        try:
            # INPUT room_number from the User
            room_number = int(input("Enter Room Number To Deallocate: "))

            # IF User not enter room_number THEN RAISE Value Error
            if not room_number:
                raise ValueError("Room Number Cannot be Empty.")

            # Check IF room_number in allocation 
            if room_number in allocation:
                # INPUT duration_of_stay from the User
                duration_of_stay = int(input("Number of Days Stay: "))

                # IF User input duration_of_stay less than 0 or zero then RAISE Value Error
                if duration_of_stay <= 0:
                    raise ValueError("Duration of Stay must be greater than zero")
            
                # Get room_prize from the rooms LIST
                room_prize = next((room["room_prize"] for room in rooms if room["room_number"] == room_number), None)

                # IF no room_prize THEN RAISE Key Error
                if room_prize is None:
                    raise KeyError("Room Prize not found for the room")
                    
                # OverFlowError Testing
                if duration_of_stay >= 10**2:
                    raise OverFlowError("Simulated overflow error for testing purposes")
                    
                # Bill Amount = Duration of Stay Multiply Room Prize
                bill_amount = (duration_of_stay * float(room_prize))/0

                # FOR room in rooms LIST
                for room in rooms:
                    # IF room_number is equal to room_number
                    if room["room_number"] == room_number:
                        # SET availability equal to available
                        room["availability"] = "available"
                        break
                # DELETE room from the allocation dictionary
                del allocation[room_number]

                # PRINT Bill Details,
                print("Room Deallocated Successfully.")
                print("------Bill Details------")
                print(f"Room Number: {room_number}")
                print(f"Number of Days Stay: {duration_of_stay}")
                print(f"Room Prize: ${room_prize} per day")
                print(f"------Bill Amount: ${bill_amount}------")

            # ELSE RAISE Key Error and PRINT the MESSAGE
            else:
                raise KeyError("Room Not Found In Allocation Records.")

        # EXCEPT RAISE Value Error and PRINT MESSAGE
        except ValueError as e:
            print(f"Value Error: {e}")

        # EXCEPT RAISE Key Error and PRINT MESSAGE
        except KeyError as e:
            print(f"Key Error: {e}")

        # EXCEPT RAISE Key Error and PRINT MESSAGE
        except OverflowError as e:
            print(f"Overflow Error: {e}")

        # EXCEPT RAISE ZeroDivisionError and PRINT MESSAGE
        except ZeroDivisionError as e:
            print(f"Zero Division Error: {e}")

    # Main function to handle menu options
    def main():
        try:
            # Create EMPTY rooms LIST
            rooms = []
            # Create EMPTY room Dictionary
            allocation = {}
            # Create filename and txt file
            filename = "LHMS_764708022.txt"

            while True:
                # PRINT HOTEL MENU OPTIONS
                print("\n----- LANGHAM HOTEL MANAGEMENT SYSTEM -----")
                print("              --- MENU ---                 ")
                print("1. Add Room")
                print("2. Delete Room")
                print("3. Display Rooms Details")
                print("4. Allocate Room")
                print("5. Display Room Allocation Details")
                print("6. Billing & De-Allocation")
                print("7. Save Room Allocation Details To File")
                print("8. Display Room Allocation Details From File")
                print("9. Backup Data")
                print("0. Exit Application")

                # PROMPT the user to Input Choice
                choice = input("Enter your choice: ")

                # IF choice equal to 1, CALL Add_Room FUNCTION
                if choice == '1':
                    Add_Room(rooms)
                # IF choice equal to 2, CALL Delete_Room FUNCTION
                elif choice == '2':
                    Delete_Room(rooms)
                # IF choice equal to 3, CALL Display_Room_Details FUNCTION
                elif choice == '3':
                    Display_Rooms_Details(rooms)
                # IF choice equal to 4, CALL Allocate_Room FUNCTION
                elif choice == '4':
                    Allocate_Room(rooms, allocation)
                # IF choice equal to 5, CALL Display_Allocation_Details FUNCTION
                elif choice == '5':
                    Display_Allocation_Details(allocation)
                # IF choice equal to 6, CALL Billing_De_Allocation FUNCTION
                elif choice == '6':
                    Billing_De_Allocation(rooms, allocation)
                # IF choice equal to 7, CALL Save_Room_Allocation_Details_To_File FUNCTION
                elif choice == '7':
                    Save_Room_Allocation_Details_To_File(filename, allocation)
                # IF choice equal to 8, CALL Display_Room_Allocation_Details_From_File FUNCTION
                elif choice == '8':
                    Display_Room_Allocation_Details_From_File(filename)
                # IF choice equal to 9, CALL Backup_File FUNCTION
                elif choice == '9':
                    Backup_File(filename)
                # IF choice equal to 0, Exiting The Application BREAK
                elif choice == '0':
                    print("Exiting The Application....")
                    break
                # ELSE PRINT "Invalid choice..." MESSAGE
                else:
                    print("Invalid Choice. Please Try Again.")
    
        except (SyntaxError, ValueError, ZeroDivisionError, IndexError, NameError, TypeError, OverflowError, IOError, ImportError, EOFError, FileNotFoundError) as e:
            print(f"An Error Occurred: {e}")


    # CALLING MAIN FUNCTION
    if __name__ == "__main__":
        main()
```
